**Q9.What will be Average Price of Vistara airline for a flight from Delhi to Hydrabad in Business Class?**

```python
filtered_df = df[(df['airline'] == 'Vistara') &
                 (df['source_city'] == 'Delhi') &
                 (df['destination_city'] == 'Hyderabad') &
                 (df['class'] == 'Business')]

average_price = filtered_df['price'].mean()
print(f"The average price of Vistara airline for a flight from Delhi to Hyderabad in Business Class is: {average_price:.2f}")
```
